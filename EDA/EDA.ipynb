# Univariate Data analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
sns.countplot(df['Embarked'])
```

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=5)
```
